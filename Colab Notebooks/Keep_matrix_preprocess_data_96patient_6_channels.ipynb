```
# 1-3 delta   4-7 ceta 8-10 lowalfa 10-12highalfa  13-21beta1 22-30beta2 
```

```
print(input1.shape)
all_data = np.concatenate((input1, input2, input3, input4),axis = 0)
print(all_data.shape)
```

```
def compress():  # 将 30个hz频道 变成6个 频道
  L =[]
 
  for i in range(96):
    New = []
    Delta = 0
    Ceta = 0
    Lowalfa = 0
    Highalfa = 0
    Beta1 = 0
    Beta2 = 0
    for j in range(0,3):
      Delta = Delta + all_data[i,j,:,:]
    Delta = Delta/3
    for j in range(3,7):
      Ceta = Ceta + all_data[i,j,:,:]
    Ceta = Ceta/4
    for j in range(7,10):
      Lowalfa = Lowalfa + all_data[i,j,:,:]
    Lowalfa = Lowalfa/3
    for j in range(10,12):
      Highalfa = Highalfa + all_data[i,j,:,:]
    Highalfa = Highalfa/2
    for j in range(12,21):
      Beta1 = Beta1 + all_data[i,j,:,:]
    Beta1 = Beta1/9
    for j in range(21,30):
      Beta2 = Beta2 + all_data[i,j,:,:]
    Beta2 = Beta2/9
    New.append(Delta)
    New.append(Ceta)
    New.append(Lowalfa)
    New.append(Highalfa)
    New.append(Beta1)
    New.append(Beta2)
    L.append(New)
  New = np.array(L)
  return New
```

```
all_data =compress()
```
